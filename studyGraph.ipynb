```python
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: set a style for cleaner visuals
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='study_hours', y='test_score', hue='mood', s=100)
plt.title("Study Hours vs Test Score (by Mood)")
plt.xlabel("Study Hours")
plt.ylabel("Test Score")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.lineplot(data=df, x='sleep_hours', y='test_score', marker='o')
plt.title("Sleep Hours vs Test Score")
plt.xlabel("Sleep Hours")
plt.ylabel("Test Score")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='mood', palette='pastel')
plt.title("Mood Distribution")
plt.xlabel("Mood")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(df[['study_hours','sleep_hours','test_score']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Between Variables")
plt.show()
```

```python
avg_study = df['study_hours'].mean()
avg_sleep = df['sleep_hours'].mean()
avg_score = df['test_score'].mean()

print(f"Average Study Hours per Day: {avg_study:.2f}")
print(f"Average Sleep Hours per Night: {avg_sleep:.2f}")
print(f"Average Test Score: {avg_score:.2f}")
```
